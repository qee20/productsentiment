```python
import pandas as pd
import glob

brand = input("What to clean now?") 

paths = glob.glob(f"../data/raw/{brand}_*.csv")

dfs = [pd.read_csv(p) for p in paths]
ytcomment_df = pd.concat(dfs, ignore_index=True)

ytcomment_df = ytcomment_df[['comment']].copy()
ytcomment_df.shape
```

```python
import re
import string
import emoji
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'http\S+|www\.\S+', ' ', text)
    text = re.sub(r'\d+', ' ', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = emoji.replace_emoji(text, replace='')
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
ytcomment_df['clean_comment'] = ytcomment_df['comment'].apply(clean_text)

ytcomment_df[['comment', 'clean_comment']].head(10)
```

```python
ytcomment_df = ytcomment_df[['clean_comment']].copy()
ytcomment_df.head()
```

```python
sebelum = len(ytcomment_df)
# cleaning
ytcomment_df['clean_comment'] = ytcomment_df['clean_comment'].apply(clean_text)

# hapus NaN
ytcomment_df = ytcomment_df.dropna(subset=['clean_comment'])

# hapus string kosong
ytcomment_df = ytcomment_df[ytcomment_df['clean_comment'].str.strip() != '']

# reset index
ytcomment_df = ytcomment_df.reset_index(drop=True)
sesudah = len(ytcomment_df)
print("Terbuang:", sebelum - sesudah)
```

```python
labels = input("What's this label for?")

ytcomment_df.to_csv(f"../data/processed/{labels}cleanedytcomment.csv", index=False)
print(f"Saved to ../data/processed/{labels}cleanedytcomment.csv")
```
