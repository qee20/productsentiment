```python
# 1. Cek baris yang kosong di kolom 'comment'
ytcomment_df['comment'].isna().sum()
# Sampel baris kosong
ytcomment_df[ytcomment_df['comment'].isna()].head()
# 2. Hitung berapa yang kosong
n_kosong = ytcomment_df['comment'].isna().sum()
print("Jumlah baris kosong di kolom 'comment':", n_kosong)
# 3. Hapus baris yang kosong di kolom 'comment' dan update ke ytcomment_df
ytcomment_df = ytcomment_df.dropna(subset=['comment'])
# 4. Cek lagi untuk memastikan sudah tidak ada yang kosong
print("Setelah dibersihkan, sisa baris:", len(ytcomment_df))
print("Jumlah baris kosong di kolom 'comment' sekarang:",
      ytcomment_df['comment'].isna().sum())
```

```python
# Memastikan kolom comment bertipe string
ytcomment_df['comment'] = ytcomment_df['comment'].astype(str)

mask_kosong_or_ws = ytcomment_df['comment'].str.strip() == ""

# Lihat berapa banyak
print("Jumlah baris kosong / whitespace-only:",
      mask_kosong_or_ws.sum())

# Kalau mau lihat contoh barisnya
ytcomment_df[mask_kosong_or_ws].head()
```

```python
# Buang baris yang kosong / whitespace-only di kolom 'comment'
ytcomment_df = ytcomment_df[~mask_kosong_or_ws].copy()

print("Sisa baris setelah buang kosong/whitespace:", len(ytcomment_df))
```

```python
import re
import string
```

```python
def clean_text(text):
    # pastikan tipe string
    text = str(text)
    
    # 1) lowercase
    text = text.lower()
    
    # 2) buang URL (http, https, www, dll)
    text = re.sub(r'http\S+|www\.\S+', ' ', text)
    
    # 3) buang angka
    text = re.sub(r'\d+', ' ', text)
    
    # 4) buang tanda baca
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # 5) buang extra whitespace (spasi berulang, dll)
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python
ytcomment_df['clean_comment'] = ytcomment_df['comment'].apply(clean_text)

ytcomment_df[['comment', 'clean_comment']].head(10)
```

```python
ytcomment_df = ytcomment_df[['clean_comment']].copy()
ytcomment_df.head()
```

```python
ytcomment_df.to_csv("../datasets/cleanedytcomment.csv", index=False)
print("Saved to ../datasets/cleanedytcomment.csv")
```
