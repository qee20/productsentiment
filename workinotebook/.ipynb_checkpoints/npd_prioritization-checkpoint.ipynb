```python
import pandas as pd
import json
import numpy as np
from datetime import datetime, timedelta

print("🚀 PRODUCT ROADMAP PRIORITIZATION FROM KANO ANALYSIS")
print("=" * 70)

# ==================== 1. LOAD KANO RESULTS ====================
print("\n📁 Loading Kano classification results...")
try:
    kano_df = pd.read_csv("kano_classification_results.csv")
    print(f"✅ Loaded Kano data: {len(kano_df)} aspects")
except FileNotFoundError:
    print("❌ File 'kano_classification_results.csv' not found.")
    print("📝 Creating sample data for demonstration...")
    
    # Sample data jika file tidak ada
    kano_df = pd.DataFrame({
        'aspect': ['Battery', 'Camera', 'Display', 'Performance', 'Price', 
                  'Design', 'Software', 'Audio', 'Security', 'Network'],
        'kano_category': ['MUST-BE', 'ONE-DIMENSIONAL', 'ONE-DIMENSIONAL', 
                         'ONE-DIMENSIONAL', 'MUST-BE', 'ATTRACTIVE', 
                         'MIXED', 'INDIFFERENT', 'MUST-BE', 'INDIFFERENT'],
        'total_mentions': [450, 380, 320, 290, 280, 210, 180, 150, 120, 90],
        'positive_pct': [15.2, 58.7, 52.3, 48.9, 12.5, 78.4, 42.3, 25.6, 8.7, 20.3],
        'negative_pct': [72.5, 35.2, 38.4, 41.2, 75.3, 8.7, 45.6, 20.1, 85.4, 15.2],
        'neutral_pct': [12.3, 6.1, 9.3, 9.9, 12.2, 12.9, 12.1, 54.3, 5.9, 64.5]
    })

print("\n📊 Kano Distribution:")
category_counts = kano_df['kano_category'].value_counts()
for category, count in category_counts.items():
    pct = (count / len(kano_df)) * 100
    print(f"  {category:20}: {count:2d} aspects ({pct:5.1f}%)")

# ==================== 2. BUSINESS CONSTRAINTS ====================
print("\n⚙️  Setting business constraints...")

# Anda bisa adjust parameter ini sesuai kondisi perusahaan
business_constraints = {
    'team_size': 10,           # Jumlah engineer
    'budget_quarters': 4,      # Timeline dalam quarters
    'resources_per_quarter': {
        'CRITICAL': 6,         # Engineer untuk critical features
        'HIGH': 3,             # Engineer untuk high priority
        'MEDIUM': 1,           # Engineer untuk medium priority
        'LOW': 0               # Engineer untuk low priority
    },
    'development_speed': {
        'MUST-BE': 1,          # Quarter untuk MUST-BE features
        'ONE-DIMENSIONAL': 2,  # Quarter untuk ONE-DIMENSIONAL
        'ATTRACTIVE': 1.5,     # Quarter untuk ATTRACTIVE
        'INDIFFERENT': 0.5     # Quarter untuk INDIFFERENT (eval only)
    }
}

print(f"  Team size: {business_constraints['team_size']} engineers")
print(f"  Timeline: {business_constraints['budget_quarters']} quarters")
print(f"  Development speed:")
for category, speed in business_constraints['development_speed'].items():
    print(f"    - {category}: {speed} quarters per major feature")

# ==================== 3. PRIORITIZATION ALGORITHM ====================
print("\n🎯 Calculating feature priorities...")

def calculate_feature_priority(row):
    """
    Calculate priority score for each feature based on Kano category and impact
    """
    base_scores = {
        'MUST-BE': 100,
        'ONE-DIMENSIONAL': 80,
        'ATTRACTIVE': 60,
        'MIXED': 40,
        'INDIFFERENT': 20,
        'REVERSE': 0
    }
    
    # Base score from Kano category
    priority_score = base_scores.get(row['kano_category'], 50)
    
    # Adjust by negative impact (higher negative = more urgent)
    if row['kano_category'] == 'MUST-BE':
        priority_score += row['negative_pct'] * 0.5
    
    # Adjust by positive impact for attractive features
    if row['kano_category'] == 'ATTRACTIVE':
        priority_score += row['positive_pct'] * 0.3
    
    # Adjust by frequency (more mentions = more important)
    priority_score += np.log1p(row['total_mentions']) * 10
    
    return priority_score

# Calculate priority scores
kano_df['priority_score'] = kano_df.apply(calculate_feature_priority, axis=1)
kano_df['priority_rank'] = kano_df['priority_score'].rank(ascending=False, method='dense').astype(int)

# Sort by priority
kano_df = kano_df.sort_values('priority_score', ascending=False)

print("\n📈 TOP 10 PRIORITY FEATURES:")
print("=" * 80)
print(f"{'Rank':>4} {'Aspect':20} {'Kano':20} {'Score':>8} {'Neg%':>6} {'Pos%':>6}")
print("-" * 80)

for idx, row in kano_df.head(10).iterrows():
    print(f"{row['priority_rank']:4d} {row['aspect'][:18]:20} {row['kano_category'][:18]:20} "
          f"{row['priority_score']:8.1f} {row['negative_pct']:6.1f} {row['positive_pct']:6.1f}")

# ==================== 4. GENERATE ROADMAP ====================
print("\n🗺️  Generating product roadmap...")

def generate_roadmap(kano_df, constraints):
    """
    Generate quarterly roadmap based on priorities and constraints
    """
    roadmap = {}
    current_quarter = 1
    
    # Calculate quarters needed based on development speed
    features_planned = []
    
    for quarter in range(1, constraints['budget_quarters'] + 1):
        quarter_name = f"Q{quarter} (Month {((quarter-1)*3)+1}-{quarter*3})"
        roadmap[quarter_name] = {
            "focus": "",
            "objectives": [],
            "features": [],
            "resources_required": 0,
            "success_metrics": []
        }
    
    # Q1: Focus on MUST-BE (Critical fixes)
    must_be_features = kano_df[kano_df['kano_category'] == 'MUST-BE'].head(3)
    if len(must_be_features) > 0:
        q1_features = []
        for _, row in must_be_features.iterrows():
            q1_features.append({
                "feature": row['aspect'],
                "priority": "CRITICAL",
                "objective": f"Reduce negative mentions from {row['negative_pct']:.1f}% to <30%",
                "resources": 2,  # Engineers needed
                "success_metric": f"Customer complaints ↓ {row['negative_pct'] - 30:.1f}%"
            })
        
        roadmap[f"Q1 (Month 1-3)"]["focus"] = "CRITICAL FIXES & STABILIZATION"
        roadmap[f"Q1 (Month 1-3)"]["features"] = q1_features
        roadmap[f"Q1 (Month 1-3)"]["resources_required"] = len(q1_features) * 2
        roadmap[f"Q1 (Month 1-3)"]["objectives"] = [
            "Eliminate major pain points",
            "Meet basic customer expectations",
            "Stabilize product performance"
        ]
        roadmap[f"Q1 (Month 1-3)"]["success_metrics"] = [
            "≥30% reduction in negative mentions for MUST-BE features",
            "Customer satisfaction score increase by 20 points",
            "Reduction in support tickets"
        ]
    
    # Q2: Focus on ONE-DIMENSIONAL (Competitive advantage)
    one_dim_features = kano_df[kano_df['kano_category'] == 'ONE-DIMENSIONAL'].head(3)
    if len(one_dim_features) > 0:
        q2_features = []
        for _, row in one_dim_features.iterrows():
            current_gap = row['positive_pct'] - row['negative_pct']
            target_gap = current_gap + 25  # Aim to improve by 25 points
            
            q2_features.append({
                "feature": row['aspect'],
                "priority": "HIGH",
                "objective": f"Increase satisfaction gap from {current_gap:+.1f} to {target_gap:+.1f}",
                "resources": 1.5,
                "success_metric": f"Satisfaction gap ↑ {25:.0f} points"
            })
        
        roadmap[f"Q2 (Month 4-6)"]["focus"] = "COMPETITIVE ADVANTAGE"
        roadmap[f"Q2 (Month 4-6)"]["features"] = q2_features
        roadmap[f"Q2 (Month 4-6)"]["resources_required"] = len(q2_features) * 1.5
        roadmap[f"Q2 (Month 4-6)"]["objectives"] = [
            "Improve key performance metrics",
            "Catch up with competitors",
            "Build reliable core features"
        ]
        roadmap[f"Q2 (Month 4-6)"]["success_metrics"] = [
            "Positive-Negative gap improvement ≥20 points",
            "Market share increase by 5%",
            "Improved review scores"
        ]
    
    # Q3: Focus on ATTRACTIVE (Differentiation)
    attractive_features = kano_df[kano_df['kano_category'] == 'ATTRACTIVE'].head(2)
    if len(attractive_features) > 0:
        q3_features = []
        for _, row in attractive_features.iterrows():
            q3_features.append({
                "feature": row['aspect'],
                "priority": "MEDIUM",
                "objective": f"Maintain high positive rate ({row['positive_pct']:.1f}%) and add innovative features",
                "resources": 1,
                "success_metric": f"Feature adoption rate ≥60%"
            })
        
        roadmap[f"Q3 (Month 7-9)"]["focus"] = "DIFFERENTIATION & INNOVATION"
        roadmap[f"Q3 (Month 7-9)"]["features"] = q3_features
        roadmap[f"Q3 (Month 7-9)"]["resources_required"] = len(q3_features) * 1
        roadmap[f"Q3 (Month 7-9)"]["objectives"] = [
            "Create unique selling points",
            "Delight customers with innovative features",
            "Build brand loyalty"
        ]
        roadmap[f"Q3 (Month 7-9)"]["success_metrics"] = [
            "Feature adoption rate ≥60%",
            "Positive social media mentions increase by 30%",
            "Customer referral rate increase"
        ]
    
    # Q4: Focus on optimization and next cycle
    roadmap[f"Q4 (Month 10-12)"]["focus"] = "OPTIMIZATION & NEXT GENERATION"
    roadmap[f"Q4 (Month 10-12)"]["features"] = [{
        "feature": "Technical debt & optimization",
        "priority": "MEDIUM",
        "objective": "Code refactoring, performance optimization, bug fixes",
        "resources": 2,
        "success_metric": "System performance improvement by 20%"
    }]
    roadmap[f"Q4 (Month 10-12)"]["resources_required"] = 2
    roadmap[f"Q4 (Month 10-12)"]["objectives"] = [
        "Technical excellence",
        "Prepare for next product generation",
        "Process improvement"
    ]
    roadmap[f"Q4 (Month 10-12)"]["success_metrics"] = [
        "Reduced bug rate by 40%",
        "Improved development velocity",
        "Technical debt reduction"
    ]
    
    return roadmap

# Generate roadmap
product_roadmap = generate_roadmap(kano_df, business_constraints)

# ==================== 5. DISPLAY ROADMAP ====================
print("\n" + "="*80)
print("📅 PRODUCT ROADMAP PRIORITIZATION")
print("="*80)

for quarter, details in product_roadmap.items():
    print(f"\n{quarter}")
    print(f"Focus: {details['focus']}")
    print(f"Resources required: {details['resources_required']} engineers")
    
    print(f"\nObjectives:")
    for obj in details['objectives']:
        print(f"  • {obj}")
    
    print(f"\nFeatures to develop:")
    for feature in details['features']:
        print(f"  • {feature['feature']} [{feature['priority']}]")
        print(f"    Objective: {feature['objective']}")
        print(f"    Success: {feature['success_metric']}")
    
    print(f"\nSuccess metrics:")
    for metric in details['success_metrics']:
        print(f"  ✓ {metric}")
    
    print("-" * 80)

# ==================== 6. RESOURCE ALLOCATION ANALYSIS ====================
print("\n" + "="*80)
print("👥 RESOURCE ALLOCATION ANALYSIS")
print("="*80)

total_resources_needed = sum(details['resources_required'] for details in product_roadmap.values())
team_size = business_constraints['team_size']

print(f"\nTotal engineers needed: {total_resources_needed:.1f}")
print(f"Available engineers: {team_size}")
print(f"Utilization: {(total_resources_needed/team_size)*100:.1f}%")

if total_resources_needed > team_size:
    print("⚠️  WARNING: Resource overallocation!")
    print(f"   Need {total_resources_needed - team_size:.1f} more engineers")
    print("   Recommendations:")
    print("   1. Extend timeline for some features")
    print("   2. Prioritize MUST-BE features first")
    print("   3. Consider outsourcing non-critical features")
else:
    print("✅ Resource allocation is feasible")
    print(f"   {team_size - total_resources_needed:.1f} engineers available for additional tasks")

# ==================== 7. RISK ASSESSMENT ====================
print("\n" + "="*80)
print("⚠️  RISK ASSESSMENT")
print("="*80)

risks = []

# Check for high negative features
high_negative = kano_df[kano_df['negative_pct'] > 60]
if len(high_negative) > 0:
    risks.append({
        "risk": "High dissatisfaction features",
        "impact": "HIGH",
        "mitigation": "Address in Q1 as CRITICAL priority",
        "features": high_negative['aspect'].tolist()[:3]
    })

# Check for low total mentions (might be niche features)
low_mentions = kano_df[kano_df['total_mentions'] < 20]
if len(low_mentions) > 0:
    risks.append({
        "risk": "Low awareness features",
        "impact": "MEDIUM",
        "mitigation": "Validate with additional market research",
        "features": low_mentions['aspect'].tolist()[:3]
    })

# Check for mixed sentiment (hard to please)
mixed_sentiment = kano_df[(kano_df['positive_pct'] > 30) & (kano_df['negative_pct'] > 30)]
if len(mixed_sentiment) > 0:
    risks.append({
        "risk": "Polarizing features",
        "impact": "MEDIUM",
        "mitigation": "Consider A/B testing different implementations",
        "features": mixed_sentiment['aspect'].tolist()[:3]
    })

for i, risk in enumerate(risks, 1):
    print(f"\n{i}. {risk['risk']} [{risk['impact']} impact]")
    print(f"   Features: {', '.join(risk['features'])}")
    print(f"   Mitigation: {risk['mitigation']}")

# ==================== 8. SAVE ALL OUTPUTS ====================
print("\n💾 Saving all outputs...")

# 8.1. Save prioritized features
prioritized_features_path = "../data/usefuldict/prioritized_features.csv"
kano_df[['priority_rank', 'aspect', 'kano_category', 'total_mentions', 
         'positive_pct', 'negative_pct', 'priority_score']].to_csv(
    prioritized_features_path, index=False, encoding='utf-8-sig')
print(f"✅ Prioritized features saved to: {prioritized_features_path}")

# 8.2. Save roadmap in JSON
roadmap_json_path = "../data/usefuldict/product_roadmap_detailed.json"
with open(roadmap_json_path, 'w', encoding='utf-8') as f:
    json.dump(product_roadmap, f, ensure_ascii=False, indent=2)
print(f"✅ Detailed roadmap saved to: {roadmap_json_path}")

# 8.3. Save executive summary
executive_summary_path = "../data/usefuldict/product_roadmap_executive_summary.txt"
with open(executive_summary_path, 'w', encoding='utf-8') as f:
    f.write("PRODUCT ROADMAP PRIORITIZATION - EXECUTIVE SUMMARY\n")
    f.write("=" * 70 + "\n\n")
    
    f.write("1. STRATEGIC FOCUS BY QUARTER:\n")
    f.write("-" * 40 + "\n")
    for quarter, details in product_roadmap.items():
        f.write(f"\n{quarter}: {details['focus']}\n")
        for feature in details['features'][:2]:  # Top 2 features per quarter
            f.write(f"  • {feature['feature']}: {feature['objective'][:60]}...\n")
    
    f.write("\n\n2. TOP 5 PRIORITY FEATURES:\n")
    f.write("-" * 40 + "\n")
    for idx, row in kano_df.head(5).iterrows():
        f.write(f"{row['priority_rank']}. {row['aspect']} ({row['kano_category']})\n")
        f.write(f"   Impact: {row['total_mentions']} mentions, "
                f"{row['negative_pct']:.1f}% negative, {row['positive_pct']:.1f}% positive\n")
    
    f.write("\n\n3. RESOURCE REQUIREMENTS:\n")
    f.write("-" * 40 + "\n")
    f.write(f"Total engineers needed: {total_resources_needed:.1f}\n")
    f.write(f"Available engineers: {team_size}\n")
    f.write(f"Utilization rate: {(total_resources_needed/team_size)*100:.1f}%\n")
    
    if total_resources_needed > team_size:
        f.write("\n⚠️  ACTION REQUIRED: Need additional resources\n")
    
    f.write("\n\n4. KEY RISKS & MITIGATIONS:\n")
    f.write("-" * 40 + "\n")
    for i, risk in enumerate(risks[:3], 1):
        f.write(f"{i}. {risk['risk']}\n")
        f.write(f"   Mitigation: {risk['mitigation']}\n")
    
    f.write("\n\n5. RECOMMENDED NEXT ACTIONS:\n")
    f.write("-" * 40 + "\n")
    f.write("1. Approve Q1 resource allocation immediately\n")
    f.write("2. Begin requirement gathering for MUST-BE features\n")
    f.write("3. Schedule stakeholder review in 2 weeks\n")
    f.write("4. Set up success metrics tracking system\n")

print(f"✅ Executive summary saved to: {executive_summary_path}")

# ==================== 9. FINAL RECOMMENDATIONS ====================
print("\n" + "="*80)
print("🎯 FINAL RECOMMENDATIONS FOR NPD")
print("="*80)

print("\n🚀 IMMEDIATE ACTIONS (Next 30 days):")
print("1. Allocate resources for Q1 MUST-BE features")
print("2. Create detailed specifications for:")
for feature in product_roadmap["Q1 (Month 1-3)"]["features"]:
    print(f"   • {feature['feature']}")
print("3. Set up success metrics tracking")

print("\n📊 MONITORING METRICS:")
print("• Weekly: Negative mention reduction progress")
print("• Monthly: Customer satisfaction scores")
print("• Quarterly: Feature adoption rates")

print("\n🔄 ITERATION CYCLE:")
print("• Review roadmap progress monthly")
print("• Adjust priorities based on new customer feedback")
print("• Re-run Kano analysis after major releases")

print("\n" + "="*80)
print("✅ PRODUCT ROADMAP PRIORITIZATION COMPLETE!")
print("="*80)

print(f"\n📁 OUTPUT FILES:")
print(f"1. prioritized_features.csv - Ranked features with scores")
print(f"2. product_roadmap_detailed.json - Detailed quarterly plan")
print(f"3. product_roadmap_executive_summary.txt - Summary for stakeholders")

print(f"\n🎯 NEXT STEPS:")
print(f"1. Present executive summary to management")
print(f"2. Allocate resources for Q1 features")
print(f"3. Begin requirement gathering and planning")
```
